```python
# Data Analyze
RootDir='/home/lihaowei/Documents/Git/HololensWorkingUI/UI/SensorData/TestData/Test2'
OutDir='/home/lihaowei/Documents/Git/HololensWorkingUI/UI/SensorData/MatlabData/Data/Test_dist_influ.mat'
TestDir=os.listdir(RootDir)
NameList=[]
# Test*frames*8*3
XYZS=[]
XYDS=[]
for f in tqdm(range(len(TestDir))):
    FileName=TestDir[f]
    if "Test" in FileName:
        PathThis=os.path.join(RootDir,FileName)
        TestThisFrame=np.load(PathThis,allow_pickle=True)
        Refrence=TestThisFrame[0][0].Points
        FrameNum=len(TestThisFrame)
        if any([(len(Fr[0].Points)!=8) for Fr in TestThisFrame]):
            continue
        All_xyd=[]
        All_xyz=[]
        for i in range(FrameNum):
            ThisFr=TestThisFrame[i][0].Points
            ThisFrxyd=TestThisFrame[i][0].xyd_Data
            xyd_fr=[]
            xyz_fr=[]
            for j in range(8):
                # Find the nearest Point
                Dist=np.sqrt(np.sum((ThisFr-Refrence[j])**2,1))
                minDist=min(Dist)
                if minDist>8:
                    # print(FileName+"   "+str(i)+"    "+str(j))
                    break
                IDThis=np.where(Dist==minDist)[0][0]
                xyd_fr.append(ThisFrxyd[IDThis])
                xyz_fr.append(ThisFr[IDThis])
            if len(xyd_fr)==8 and len(xyz_fr)==8:
                All_xyd.append(xyd_fr)
                All_xyz.append(xyz_fr)
        if len(All_xyz)==len(TestThisFrame):
            NameList.append(FileName)
            XYZS.append(All_xyz)
            XYDS.append(All_xyd)
Position_xyz=[]
std_xyz=[]
Position_xyd=[]
std_xyd=[]
Err_all=[]
for frame in tqdm(range(len(XYDS))):
    for point in range(8):
        ThisTest_xyz=XYZS[frame]
        ThisTest_xyd=XYDS[frame]
        xyz_pts=[P[point] for P in ThisTest_xyz]
        xyd_pts=[P[point] for P in ThisTest_xyd]
        mean_xyz=np.mean(xyz_pts,0)
        Err_all.append(np.mean(np.sqrt(np.sum((xyz_pts-mean_xyz)**2,1))))
        Position_xyd.append(np.mean(xyd_pts,0))
        Position_xyz.append(np.mean(xyz_pts,0))
        std_xyd.append(np.std(xyd_pts,0))
        std_xyz.append(np.std(xyz_pts,0))
savemat(OutDir,{"xyz":np.array(Position_xyz),"xyd":np.array(Position_xyd),
                                "xyz_std":np.array(std_xyz),"xyd_std":np.array(std_xyd),"Err_all":Err_all})
```

**Input** : A normal frame AHAT IMG, `type = AHATFrame`
**Output** : Position of every pixel on unit plane, `type = np.array([512,512],dtype='float')`

```python
from My_classes import *
import numpy as np
import os
from scipy.io import savemat,loadmat
import matplotlib
import matplotlib.pyplot as plt
import cv2
Intrincs=loadmat('AHAT_Para_Python.mat')
mtx=Intrincs['Mtx']
distcoef=Intrincs['dist']
AHATDataDemo=np.load("CameraData/AHATRecordData.npy",allow_pickle=True)
AHATFrame=AHATDataDemo[0]
DepthImg=AHATFrame.MatDepth
RfImg=AHATFrame.MatReflectivity
RfImgcopy=RfImg
RfImgcopy[RfImgcopy>0]=1
RfImgcopy=np.array(1-RfImgcopy,dtype='uint8')
plt.imshow(RfImgcopy,cmap='gray')
retval,labels,stats,centroids=cv2.connectedComponentsWithStats(RfImgcopy,connectivity=8)
```

```python
print(labels)
```

```python
invalidlabel=((labels==1)+(labels==1153))>0
validlabel=1-invalidlabel
plt.imshow(validlabel,cmap='gray')
```

```python
import cv2
import numpy as np
import glob
from tqdm import tqdm
from scipy.io import savemat, loadmat
import matplotlib.pyplot as plt
import os
X_position=np.zeros([512,512],dtype='float')
Y_position=np.zeros([512,512],dtype='float')
for i in tqdm(range(512)):
    for j in range(512):
        if validlabel[i,j]>0:
            PointXY=cv2.undistortPoints(np.array([j,i],dtype='float'),mtx,distcoef)[0][0]
            X_position[i,j]=PointXY[0]
            Y_position[i,j]=PointXY[1]
```

```python
savemat("AHATXYData.mat",{"X_position":X_position,"Y_position":Y_position})
```

```python
plt.imshow(np.abs(X_position),cmap='gray')
```

```python
savemat("MatlabData/XTunitplanedata.mat",{"X_position":X_position,"Y_position":Y_position})
```
