```python
# rot 10 mm 
num=20
Z_err_H=[]
Z_err_Hf=[]
Z_err_NDI=[]

Z_std_H=[]
Z_std_Hf=[]
Z_std_NDI=[]
repeat_time=10000
def_len=1
for n in range(num):
    filename="Test/PrecisionTest/20220227_test_precision_rotation_10degree_t_"+str(n+1)+".npy"
    Data=np.load(filename,allow_pickle=True).tolist()
    H_rot=[]
    NDI_rot=[]
    Hf_rot=[]
    for k in range(repeat_time):
        m=random.randint(0,99)
        n=random.randint(0,99)
        ROT1=Data['HololensPose'][0][m].Transformatrix[0][0:3,0:3]
        ROT2=Data['HololensPose'][1][n].Transformatrix[0][0:3,0:3]
        rot=np.arccos((np.trace(ROT1@np.linalg.inv(ROT2))-1)/2)*180/np.pi
        H_rot.append(rot)
        
    for k in range(repeat_time):
        m=random.randint(0,99)
        n=random.randint(0,99)
        ROT1=Data['HololensPose'][0][m].TransformatrixFiltered[0][0:3,0:3]
        ROT2=Data['HololensPose'][1][n].TransformatrixFiltered[0][0:3,0:3]
        rot=np.arccos((np.trace(ROT1@np.linalg.inv(ROT2))-1)/2)*180/np.pi
        Hf_rot.append(rot)
        
    for k in range(repeat_time):
        m=random.randint(0,99)
        n=random.randint(0,99)
        ROT1=Data['NDIPose'][0][m][0][0:3,0:3]
        ROT2=Data['NDIPose'][1][n][0][0:3,0:3]
        rot=np.arccos((np.trace(ROT1@np.linalg.inv(ROT2))-1)/2)*180/np.pi
        NDI_rot.append(rot)
                                                                                                       
    Z_err_H.append(np.mean(H_rot))
    Z_err_Hf.append(np.mean(Hf_rot))
    Z_err_NDI.append(np.mean(NDI_rot))
    Z_std_H.append(np.std(H_rot))
    Z_std_Hf.append(np.std(Hf_rot))
    Z_std_NDI.append(np.std(NDI_rot))
plt.figure()
plt.plot(np.array(list(range(20)))+1,np.array(Z_err_H),label="Hololens")
plt.plot(np.array(list(range(20)))+1,np.array(Z_err_Hf),label="Hololens Filtered")
plt.plot(np.array(list(range(20)))+1,np.array(Z_err_NDI),label="NDI")
plt.xlabel("Times (1-20)")
plt.xticks(np.array(list(range(20)))+1)
plt.ylabel("Distance (mm)")
plt.title("Z axis, detected mean movement, GT=1mm")
plt.legend()
plt.figure()
plt.title("Z axis, detected movement standard error")
plt.plot(np.array(list(range(20)))+1,np.array(Z_std_H),label="Hololens")
plt.plot(np.array(list(range(20)))+1,np.array(Z_std_Hf),label="Hololens Filtered")
plt.plot(np.array(list(range(20)))+1,np.array(Z_std_NDI),label="NDI")
plt.xlabel("Times (1-20)")
plt.xticks(np.array(list(range(20)))+1)
plt.ylabel("Standard Error (mm)")
plt.legend()
```

```python
savemat("./Test/Result/NDI_rot_10_degree",{"Mean_Rot":Z_err_NDI,"Std":Z_std_NDI})
savemat("./Test/Result/Hololens_rot_10_degree",{"Mean_Rot":Z_err_H,"Std":Z_std_H})
savemat("./Test/Result/HololensF_rot_10_degree",{"Mean_Rot":Z_err_Hf,"Std":Z_std_Hf})
```

```python
# zaxis 1mm GT
num=20
Z_err_H=[]
Z_err_Hf=[]
Z_err_NDI=[]

Z_std_H=[]
Z_std_Hf=[]
Z_std_NDI=[]
repeat_time=10000
def_len=1
for n in range(num):
    filename="Test/PrecisionTest/20220223_test_precision_movement_z_1mm_t_"+str(n+1)+".npy"
    Data=np.load(filename,allow_pickle=True).tolist()
    H_reses=[]
    N_reses=[]
    Hf_reses=[]
    for k in range(repeat_time):
        m=random.randint(0,99)
        n=random.randint(0,99)
        Hololens_shift=Data['HololensPose'][1][m].Transformatrix[0][0:3,3]-Data['HololensPose'][0][n].Transformatrix[0][0:3,3]
        Hololens_res=np.linalg.norm(Hololens_shift)-def_len
        H_reses.append(Hololens_res)
    for k in range(repeat_time):
        m=random.randint(0,99)
        n=random.randint(0,99)
        HololensF_shift=Data['HololensPose'][1][m].TransformatrixFiltered[0][0:3,3]-Data['HololensPose'][0][n].TransformatrixFiltered[0][0:3,3]
        HololensF_res=np.linalg.norm(HololensF_shift)-def_len
        Hf_reses.append(HololensF_res)
    for k in range(repeat_time):
        m=random.randint(0,99)
        n=random.randint(0,99)
        NDI_shift=Data['NDIPose'][0][m][0][0:3,3]-Data['NDIPose'][1][n][0][0:3,3]
        NDI_res=np.linalg.norm(NDI_shift)-def_len
        N_reses.append(NDI_res)
                                                                                                       
    Z_err_H.append(np.mean(H_reses))
    Z_err_Hf.append(np.mean(Hf_reses))
    Z_err_NDI.append(np.mean(N_reses))
    Z_std_H.append(np.std(H_reses))
    Z_std_Hf.append(np.std(Hf_reses))
    Z_std_NDI.append(np.std(N_reses))
plt.figure()
plt.plot(np.array(list(range(20)))+1,np.array(Z_err_H)+def_len,label="Hololens")
plt.plot(np.array(list(range(20)))+1,np.array(Z_err_Hf)+def_len,label="Hololens Filtered")
plt.plot(np.array(list(range(20)))+1,np.array(Z_err_NDI)+def_len,label="NDI")
plt.xlabel("Times (1-20)")
plt.xticks(np.array(list(range(20)))+1)
plt.ylabel("Distance (mm)")
plt.title("Z axis, detected mean movement, GT=1mm")
plt.legend()
plt.figure()
plt.title("Z axis, detected movement standard error")
plt.plot(np.array(list(range(20)))+1,np.array(Z_std_H),label="Hololens")
plt.plot(np.array(list(range(20)))+1,np.array(Z_std_Hf),label="Hololens Filtered")
plt.plot(np.array(list(range(20)))+1,np.array(Z_std_NDI),label="NDI")
plt.xlabel("Times (1-20)")
plt.xticks(np.array(list(range(20)))+1)
plt.ylabel("Standard Error (mm)")
plt.legend()
```
