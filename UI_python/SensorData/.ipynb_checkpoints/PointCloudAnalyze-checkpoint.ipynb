```python
# import 
from My_classes import *
import numpy as np
import os
from scipy.io import savemat,loadmat
import matplotlib.pyplot as plt
from filterpy.kalman import KalmanFilter
from filterpy.common import Q_discrete_white_noise
import cv2
# from roipoly import RoiPoly
import IPython.display as Disp
from ipywidgets import widgets
import math
import open3d
```

```python
# bbox selection
class bbox_select():
    %matplotlib notebook 
    def __init__(self,im):
        self.im = im
        self.selected_points = []
        self.fig,ax = plt.subplots()
        self.img = ax.imshow(self.im.copy())
        self.ka = self.fig.canvas.mpl_connect('button_press_event', self.onclick)
        disconnect_button = widgets.Button(description="Disconnect mpl")
        Disp.display(disconnect_button)
        disconnect_button.on_click(self.disconnect_mpl)


        
    def poly_img(self,img,pts):
        pts = np.array(pts, np.int32)
        pts = pts.reshape((-1,1,2))
        cv2.polylines(img,[pts],True,(np.random.randint(0,255),np.random.randint(0,255),np.random.randint(0,255)),2)
        return img

    def onclick(self, event):
    #display(str(event))
        self.selected_points.append([event.xdata,event.ydata])
        if len(self.selected_points)>1:
            self.fig
            self.img.set_data(self.poly_img(self.im.copy(),self.selected_points))
    def disconnect_mpl(self,_):
        self.fig.canvas.mpl_disconnect(self.ka)
```

```python
CameraCollection=np.load("/run/user/1000/gvfs/smb-share:server=192.168.249.160,share=f/Share/Lihaowei/Data/HololensPerceptionData/PerceptionData/Perception_Dynamic_Pose_1_Data_1/AHAT.npy",allow_pickle=True)
_intrin=loadmat('AHAT_Para_Python.mat')
_mtx=_intrin['Mtx']
_distcoef=_intrin['dist']
_xyrecon=loadmat('AHATXYData.mat')
_X_unit=_xyrecon['X_position']
_Y_unit=_xyrecon['Y_position']
```

```python
# run image selection
bs = bbox_select(CameraCollection[0].MatDepth)
```

```python
# display roi
arr = np.array([bs.selected_points],'int')
mask = cv2.fillPoly(np.zeros([512,512],np.uint8),arr,1)
mask=np.array(mask)
plt.imshow(mask)
```

```python
# mean and 
roi_pixels=np.array(np.where(mask)).T
uvd=np.zeros([roi_pixels.shape[0],3])
XYUnit=np.zeros([roi_pixels.shape[0],2])
XYZ=np.zeros([roi_pixels.shape[0],3])
AllImagesDepth=np.array([np.double(CameraCollection[i].MatDepth) for i in range(CameraCollection.shape[0])])
ImageMean=np.mean(AllImagesDepth,0)
for i in range(roi_pixels.shape[0]):
    uvd[i,0:2]=roi_pixels[i,0:2]
    # uvd[i,2]=CameraCollection[0].MatDepth[roi_pixels[i,0],roi_pixels[i,1]]
    uvd[i,2]=ImageMean[roi_pixels[i,0],roi_pixels[i,1]]
    XYUnit[i,0]=_X_unit[roi_pixels[i,0],roi_pixels[i,1]]
    XYUnit[i,1]=_Y_unit[roi_pixels[i,0],roi_pixels[i,1]]
Dist=np.sqrt(XYUnit[:,0]**2+XYUnit[:,1]**2+1)
XYZ[:,0]=XYUnit[:,0]/Dist*uvd[:,2]
XYZ[:,1]=XYUnit[:,1]/Dist*uvd[:,2]
XYZ[:,2]=uvd[:,2]/Dist
test_pcd = open3d.geometry.PointCloud() 
test_pcd.points=open3d.utility.Vector3dVector(XYZ)
open3d.io.write_point_cloud("TestPointCloud.pcd",test_pcd)
```
